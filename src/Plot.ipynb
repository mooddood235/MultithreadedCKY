```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def unzip(path):
    xs = []
    ys = []
    with open(f'../{path}') as f:
        lines = f.readlines()
        for l in lines:
            x, y = l.split()
            xs.append(float(x))
            ys.append(float(y))
    return xs, ys
```

```python
# Test Simple Bracket

single_xs, single_ys = unzip('SingleSimpleBracket')
multi_xs, multi_ys = unzip('MultiSimpleBracket')

#plt.figure(figsize=(10, 10))

plt.plot(single_xs, single_ys, linestyle='-', label='Single')
plt.plot(multi_xs, multi_ys, linestyle='-', label='Multi')

plt.xlabel('Sentence Length')
plt.ylabel('Parse Time (Seconds)')
plt.title("Sentence Length vs. Parse Time (Fully Bracketed)")
plt.grid(True)
plt.legend()
plt.savefig('FullyBracketed.png')
plt.show()

diff = np.subtract(single_ys, multi_ys)
#plt.figure(figsize=(10, 10))
plt.plot(single_xs, diff, linestyle='-')
plt.xlabel('Sentence Length')
plt.ylabel('Time Difference')
plt.title("Sentence Length vs. Time Difference (Fully Bracketed)")
#plt.yscale('symlog')
plt.grid(True)
plt.savefig('FullyBracketedDiff.png')
plt.show()
```
